One of the most powerful and convenient features of pandas time series is time-based indexing — using dates and times to intuitively organise and access our data.

Represent the time component in the index of a Series or DataFrame so manipulations can be performed with respect to the time element.


```python
# timebased indexing
import datetime
import pandas as pd
import numpy as np


# use pandas to convert all different inputs to datetime format.
dti = pd.to_datetime(['1/1/2018', np.datetime64('2018-01-01'), datetime.datetime(2018,1,1)])
print(dti)
print("\n")

#Generate sequences of fixed-frequency dates and time spans, H is hour
dti_2 = pd.date_range('2018-01-01', periods=10, freq='H')
print(dti_2)
print("\n")

# Resampling or converting a time series to a particular frequency with index
tseries = pd.Series(range(len(dti_2)), index = dti_2)
print(tseries)
print("\n")

# resamples the time series to 2 hourly and find the mean
tseries.resample('2H').mean()
```

```python
# diff - calculates the difference between two elements
# lag - difference is the difference between two consecutive observations.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.tsa.api import VAR

# load dataset
data = pd.read_excel('SW2001_Data.xlsx')

# column: year and quarter number
# create an index based on date
data.index = pd.DatetimeIndex(data['obs'])
data_use = data[['Inflation', 'Unemployment', 'Fed Funds']]
data_use.head()

# plot graphs against time which is the index
fig, axes = plt.subplots(nrows=3, ncols=1, dpi=120, figsize=(6,6))
fig.tight_layout(pad = 3.0)

data['Inflation'].plot(title = 'Inflation', ax = axes[0])

# plot unemployment against time
data['Unemployment'].plot(title = 'Unemployment', ax = axes[1])

# plot federal funds against time
data['Fed Funds'].plot(title='Federal Funds', ax = axes[2])

# compute changes, using log will visit this more in regression, log will provide on change
# diff is the log difference
# dropna - remove missing values
data_ret = np.log(data_use).diff().dropna()
# construct model
model = VAR(data_ret)

# fit model using 8 lags
# lag order

results = model.fit(8)
print(results.summary())
print("\n")

# plot autocorrelation function
print(results.plot_acorr())
print("\n")


# forecast
lag_order = results.k_ar

print(results.forecast(data_ret.values[-lag_order:],8))
print("\n")

results.plot_forecast(10)
irf = results.irf(10)
irf.plot(orth=False)

# cumulative effect, cumulative response
irf.plot_cum_effects(orth=False)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas.plotting import lag_plot
from pandas.plotting import autocorrelation_plot
from pandas import concat
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.tsa.ar_model import AutoReg
from sklearn.metrics import mean_squared_error
from math import sqrt

# mathplotlib inline

dataset = pd.read_csv('temperature.txt', sep=",", parse_dates=['Date'], index_col = 'Date')
print(dataset.shape)
print(dataset.head())

# plot dataset
dataset.plot()
plt.show()

# plot the data on the day t and previous day, t-1
lag_plot(dataset)
plt.show()

# the examples below shows a lagged version of the minimum daily temperatures dataset
# and calculate the correlation matrix of each colum with other columns, including itself
values = pd.DataFrame(dataset.values)
dataframe = pd.concat([values.shift(1),values], axis=1)
dataframe.columns = ['t-1','t']
result = dataframe.corr()
print(result)
# note in the result, there is a strong correlation between the observation and the lag = 1
# value

# plot the correlation cofficient for each lag variable
# this can very quickly give an idea of which lag variables may be good candidtates
# for use in a predictive model
# and how the relationship between the observation and its historic values changes over time.

autocorrelation_plot(dataset)
plt.show()

# develop a persistence model
# we want to develop a model to preict the last 7 days of minimum temperatures in the dataset given all prior observations
# we will use persistence models to compare the auto regression model
# test dataset - last 7 days of the dataset
# we would like to make 7 one day forecasts
# split into train and test sets

X = dataframe.values
train, test = X[1:len(X)-7], X[len(X)-7:]
train_X, train_y = train[:,0], train[:,1]
test_X, test_y = test[:,0], test[:,1]


# persistence model
def model_persistence(x):
  return x

# walk-forward validation
predictions = list()
for x in test_X:
  yhat = model_persistence(x)
  predictions.append(yhat)
test_score = mean_squared_error(test_y, predictions)
print('Test MSE which means Mean Square Error: %.3f' %test_score)
# plot predictions vs expected
plt.plot(test_y)
plt.plot(predictions, color='red')
plt.show()
```

```python
#https://machinelearningmastery.com/autoregression-models-time-series-forecasting-python/
#Create an Autoregressive model
# create and evaluate an updated autoregressive model
import pandas as pd
from matplotlib import pyplot
from statsmodels.tsa.ar_model import AutoReg
from sklearn.metrics import mean_squared_error
from math import sqrt

# load dataset
dataset = pd.read_csv('temperature.txt', header=0, index_col=0, parse_dates=True, squeeze=True)

# split dataset
X = dataset.values
train, test = X[1:len(X)-7], X[len(X)-7:]

# train autoregression
window = 29
model = AutoReg(train, lags=29)
model_fit = model.fit()
coef = model_fit.params

# walk forward over time steps in test
history = train[len(train)-window:]
history = [history[i] for i in range(len(history))]
predictions = list()
for t in range(len(test)):
    length = len(history)
    lag = [history[i] for i in range(length-window,length)]
    yhat = coef[0]
    for d in range(window):
        yhat += coef[d+1] * lag[window-d-1]
    obs = test[t]
    predictions.append(yhat)
    history.append(obs)
    print('predicted=%f, expected=%f' % (yhat, obs))
rmse = sqrt(mean_squared_error(test, predictions))
print('Test RMSE: %.3f' % rmse)
# plot
pyplot.plot(test)
pyplot.plot(predictions, color='magenta')
pyplot.show()
```
